```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#data collection and preprocessing
movie_data = pd.read_csv('movies.csv')
```

```python
movie_data.head()
```

```python
movie_data.shape
```

```python
#4803 rows and 24 coloumns
```

```python
#selecting the needed columns  for recommendation 
selected_columns = ['genres', 'keywords','overview', 'tagline', 'cast', 'director']
selected_columns  
```

```python
#filling all the na values in these columns
for i in selected_columns:
    movie_data[i]= movie_data[i].fillna('') 
```

```python
#combining all the selected columns
combined = movie_data['genres']+' '+movie_data['keywords']+' '+movie_data['overview']+' '+movie_data['tagline']+' '+movie_data['cast']+' '+movie_data['director']
```

```python
combined
```

```python
#converting to numerical values ie to feature vectors
```

```python
column_vectors = TfidfVectorizer().fit_transform(combined)
```

```python
print(column_vectors)
```

```python
#cosine similarity :- similarity scores
similarity = cosine_similarity(column_vectors)
```

```python
similarity
```

```python
similarity.shape
```

```python
#gettinig movie name from the user
movie = input('ENTER YOUR FAVOURITE MOVIE NAME:')
```

```python
#creating list of all movies
movies_list = movie_data['title'].tolist()
```

```python
print(movies_list)
```

```python
#finding close match we use difflib here
close_match = difflib.get_close_matches(movie, movies_list)
```

```python
close_match
```

```python
closest_match = close_match[0]
```

```python
closest_match
```

```python
#finding index of the closest_match
index_closest = movie_data[movie_data.title==closest_match]['index'].values[0]
```

```python
index_closest
```

```python
#getting list of similar movies
similarity_score = list(enumerate(similarity[index_closest]))
```

```python
similarity_score
```

```python
#finding the higher similar movies by sorting in descending order
most_similar = sorted(similarity_score, key = lambda x:x[1], reverse = True)
```

```python
most_similar
```

```python
#print top similar movies based on index we find the title
print('SUGGESTED 25 MOVIES ARE:-')
j = 1
for i in most_similar:
    index = i[0]    
    title_from_index = movie_data[movie_data.index==index]['title'].values[0]
    if(j <= 25):
        print(j,'. ',title_from_index)
        j += 1
        
```

```python
#All steps together
movie = input('ENTER YOUR FAVOURITE MOVIE NAME:')#getting users favourite movie

movies_list = movie_data['title'].tolist()#creating list of movies

close_match = difflib.get_close_matches(movie, movies_list)#finding similar
if len(close_match):
   
    
    closest_match = close_match[0]#find the top most closest

    index_closest = movie_data[movie_data.title==closest_match]['index'].values[0]#finding its index

    similarity_score = list(enumerate(similarity[index_closest]))#finding its similar

    most_similar = sorted(similarity_score, key = lambda x:x[1], reverse = True)#sorting 

    print('SUGGESTED 25 MOVIES ARE:-')
    j = 1
    for i in most_similar:
        index = i[0]    
        title_from_index = movie_data[movie_data.index==index]['title'].values[0] #printing its top 25 ...with index we find the title
        if(j <= 25):
            print(j,title_from_index)
            j += 1
else:
        print("NO SIMILAR MOVIES LIKE",movie)
```
